```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys
sys.path.append('..')

import pepper
from pepper import SimulationFdfd_TE, SimulationFdfd_TM
```

### Quickly define the size of the simulation

```python
um = 1e-6
nm = 1e-9
Lx = 10 * um
Ly = 5 * um

dL = 50 * nm
```

```python
Nx = int(Lx / dL)
Ny = int(Ly / dL)

print(f"Size: {Nx} x {Ny}")
```

```python
eps = np.ones((Nx, Ny)) * 5.0
extent = [0, Lx, 0, Ly]
plt.imshow(eps.T, origin='lower', extent=extent)
plt.show()
```

```python
sim = SimulationFdfd_TE(eps, dL, [10, 10], wavelength=1550 * nm)
```

```python
src = np.zeros_like(eps)
src[Nx // 2, Ny // 2] = 1.0
```

```python
Ez = sim.solve(src)
```

```python
plt.imshow(np.real(Ez).T, extent=extent, origin='lower');
```

```python
sim_Hz = SimulationFdfd_TM(eps, dL, [10, 10], wavelength=1550 * nm)
```

```python
Hz = sim_Hz.solve(src)
```

```python
plt.imshow(np.real(Hz).T, extent=extent, origin='lower');
```
